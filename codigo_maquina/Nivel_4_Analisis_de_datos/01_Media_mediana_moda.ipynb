## Media, Mediana y Moda

- Media: consiste en obtener el promedio de los valores. Es decir, sumar la lista
de valores y el resultado dividirlo por la cantidad de datos.

- Mediana: consiste en ordenar en forma ascendente los valores, y encontrar el que se encuentra en el medio. Por ejemplo, si tuviéramos 11 datos numéricos ordenados, la mediana correspondería al valor del sexto dato. Si en cambio tenemos un número par de datos y en el medio tenemos dos datos en vez de uno, ambos se deben sumar y el resultado dividirlo por 2.

- Moda: consiste en el valor que se repite más frecuentemente en nuestro conjunto de datos.

![Media, mediana, moda](https://marielmatesblog.wordpress.com/wp-content/uploads/2017/05/estad-media-mediana-rango-moda.jpg?w=720)

```python
import pandas as pd
```

```python
goles = pd.Series([0, 3, 3, 0, 8, 1, 1, 0, 0, 1, 5])
goles
```

```python
goles.mean() # media
```

```python
goles.median() # mediana
```

```python
goles.mode() # moda
```

Los valores de la media, mediana y moda varían en función de la distribución de los datos. Gráficamente podemos hallar estas conclusiones:

![distribuciones](https://calculadorasonline.com/wp-content/uploads/2021/02/Media-mediana-y-moda.jpg)

En el ejemplo que vimos arriba con los goles, vemos que siguen una distribución sesgada positiva.
