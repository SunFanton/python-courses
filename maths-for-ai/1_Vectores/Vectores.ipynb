## Vectores en Python

<p style="font-size:1.1rem; text-align: justify">En Python, un vector puede representarse de muchas maneras, la más común y simple es mediante una lista de números normal de Python. Por ejemplo:</p>

```python
v = [10.5, 5.2, 3.25, 7.0] #Esto es una lista normal de numeros de Python
v
```

<p style="font-size:1.1rem; text-align: justify">Pero para la gran cantidad de cálculos matemáticos y científicos que se requieren en Ciencia de Datos e Inteligencia Artificial, lo más conveniente es utilizar las implementaciones de vectores y operaciones optimizadas que nos provee la librería <b>NumPy</b>, ya que Python suele ser más lento para esto. Por otro lado, NumPy corre sobre C y C++, lo que favorece a su rápidez de ejecución. Por ejemplo:</p>

```python
import numpy as np
video = np.array([10.5, 5.2, 3.25, 7.0]) # esta es la manera de generar arreglos en NumPy, de la dimension que sea
print(video) # imprime el valor del vector
video # imprime el vector en si
```

<p style="font-size:1.1rem; text-align: justify">El tamaño de un vector (cantidad de componentes o elementos del arreglo) lo puedo conocer mediante el atributo <b>size</b>:</p>

```python
video.size
```

<p style="font-size:1.1rem; text-align: justify">Para acceder a los diferentes elementos del vector, necesitamos utilizar índices. En Matemáticas, dichos índices comienzan en 1, mientras que en programación empiezan en 0:</p>

```python
video[2] # accedemos al tercer elemento del vector
```

### Vectores 2D

#### Ejemplo 1 - Usando puntos

```python
u = np.array([2, 5]) # vector 2D (de 2 componentes: x, y)
v = np.array([3, 1])
nombres = ['Punto 1', 'Punto 2']
```

```python
x_coords, y_coords = zip(u, v)
plt.scatter(x_coords, y_coords, color=["r","b"])
plt.axis([0, 9, 0, 6]) # indican los valores minimo y maximo que se mostraran en ambos ejes
# Definir el desplazamiento para los nombres
desplazamiento_x = 0.1
desplazamiento_y = 0.1
# Agregar etiquetas a cada punto
for i, nombre in enumerate(nombres):
    plt.text(x_coords[i] + desplazamiento_x, y_coords[i] + desplazamiento_y, nombre, ha='right')
plt.grid()
plt.show()
```

#### Ejemplo 2 - Usando flechas

```python
# Definimos una funcion reutilizable para poder crear vectores en forma de flecha:
'''
La funcion arrow necesita la coordenada en x y la coordenada en y del inicio (punto de aplicacion) de la flecha o vector.
Las coordenadas en x e y del final, y otras caracteristicas (color, grosor, etc)
'''
def plot_vector2d(vector2d, origin=[0,0], **options):
    return plt.arrow(origin[0], origin[1], vector2d[0], vector2d[1],
                     head_width=0.2, head_length=0.3, length_includes_head=True,
                     **options)
```

<p style="font-size:1rem">Dibujamos los vectores <b>u</b> y <b>v</b> como flechas:</p>

```python
plot_vector2d(u, color='r')
plot_vector2d(v, color='b')
nomb_vectores = ['u', 'v']
plt.axis([0, 9, 0, 6])
for i, nombre in enumerate(nomb_vectores):
    plt.text(x_coords[i] + desplazamiento_x, y_coords[i] + desplazamiento_y, nombre, ha='right')
plt.grid()
plt.show()
```

## Norma de un vector

![Norma de un vector](./images/Vectores-parte3.jpg)

#### Python puro

<p style="font-size:1rem; text-align: justify">Usando Python puro (sin Numpy):</p>

```python
def vector_norm(vector):
    squares = [element**2 for element in vector]
    return sum(squares)**0.5 # elevar a la 1/2 es lo mismo que obtener la raiz cuadrada
```

```python
print("||", u, "|| = ", vector_norm(u))
```

#### Numpy

<p style="font-size:1rem; text-align: justify">Usando Numpy (que es mucho más eficiente) con la función norm del módulo linalg (Linear Algebra):</p>

```python
import numpy.linalg as LA
LA.norm(u)
```

#### Verificamos gráficamente

```python
# Comprobemos usando el radio de un circulo que la norma del vector u sea realmente aprox 5,4
radius = LA.norm(u)
plt.gca().add_artist(plt.Circle((0,0), radius, color = "#DDDDDD"))
plot_vector2d(u, color="r")
plt.axis([0, 9, 0, 6])
plt.grid()
plt.show()
```

<p style="font-size:1rem; text-align: justify">Si se tiene un polígono definido por una cantidad de puntos (vectores) y se suma un vector v a todos estos puntos, entonces todo el polígono se traslada o se desplaza <b>v</b>. Esto es traslación geométrica:</p>

```python
t1 = np.array([2, 0.25])
t2 = np.array([2.5, 3.5])
t3 = np.array([1, 2])

x_coords, y_coords = zip(t1, t2, t3, t1)
plt.plot(x_coords, y_coords, "c--", x_coords, y_coords, "co") # c-- indica cyan y el linestyle, por ejemplo

plot_vector2d(v, t1, color="r", linestyle=":")
plot_vector2d(v, t2, color="r", linestyle=":")
plot_vector2d(v, t3, color="r", linestyle=":")

t1b = v + t1
t2b = v + t2
t3b = v + t3

x_coords_b, y_coords_b = zip(t1b, t2b, t3b, t1b)
plt.plot(x_coords_b, y_coords_b, "b-", x_coords_b, y_coords_b, "bo")

plt.text(4, 4.2, "v", color="r", fontsize=18)
plt.text(3, 2.3, "v", color="r", fontsize=18)
plt.text(3.5, 0.4, "v", color="r", fontsize=18)

plt.axis([0, 6, 0, 5])
plt.grid()
plt.show()
```

## Multiplicación por un escalar

<p style="font-size:1rem; text-align: justify">Se puede multiplicar un vector por un número o escalar. En esta operación, cada componente del vector se multiplica por el escalar, obteniendo como resultado un nuevo vector. Por ejemplo:</p>

```python
result = 1.5 * u
print("1.5 *", u, "=", result) 
```

<p style="font-size:1rem; text-align: justify">Gráficamente, la multiplicación por escalar da como resultado cambiar la escala de una figura, de ahí el nombre escalar. Por ejemplo:</p>

```python
k = 2.5
t1c = k * t1
t2c = k * t2
t3c = k * t3

plt.plot(x_coords, y_coords, "c--", x_coords, y_coords, "co")

plot_vector2d(t1, color="r")
plot_vector2d(t2, color="r")
plot_vector2d(t3, color="r")

x_coords_c, y_coords_c = zip(t1c, t2c, t3c, t1c)
plt.plot(x_coords_c, y_coords_c, "b-", x_coords_c, y_coords_c, "bo")

plot_vector2d(k * t1, color="b", linestyle=":")
plot_vector2d(k * t2, color="b", linestyle=":")
plot_vector2d(k * t3, color="b", linestyle=":")

plt.axis([0, 9, 0, 9])
plt.grid()
plt.show()
```

## Producto escalar

![Producto escalar](./images/Vectores-parte7.jpg)

### En Python

<p style="font-size:1rem; text-align: justify">El producto punto es bastante simple de implementar con Python puro:</p>

```python
print("Vector u:", u)
print("Vector v:", v)

# Para recordar una vez mas zip()
res = zip(u, v)
print("zip(u, v) =", tuple(res))

# Producto punto para vectores de dos componentes
def dot_product(v1, v2):
    return sum(v1i * v2i for v1i, v2i in zip(v1, v2))

print("Producto punto de u con v:", dot_product(u, v))
```

<p style="font-size:1rem; text-align: justify">Pero Numpy proporciona una implementación mucho más eficiente con la función dot:</p>

```python
np.dot(u, v)
```

<p style="font-size:1rem; text-align: justify">Otra forma:</p>

```python
u.dot(v)
```

<p style="font-size:1rem; text-align: justify"><b>Precaución</b>: el operador * realizará una multiplicación elemento a elemento, pero no realizará la suma final, con lo cual NO es un producto escalar:</p>

```python
print(" ", u)
print("*", v, "NO es producto punto")
print("-"*10)
print(" ", u * v) # Multiplica el primer elemento de u con el primero de v y el segundo de u con el segundo de v, dando un vector
```

### Algunas aplicaciones importantes del producto escalar

#### Cálculo del ángulo entre vectores

![Angulo entre vectores](./images/Vectores-parte9.jpg)

```python
# Calculamos el angulo entre u y v en radianes:
def vector_angle(u, v):
    cos_theta = u.dot(v)/(LA.norm(u) * LA.norm(v))
    return np.arccos(np.clip(cos_theta, -1, 1))

'''
Nota: debido a pequeños errores de coma flotante, cos_theta puede estar ligeramente fuera del intervalo [-1, 1], lo que haria que 
arccos fallara. Es por eso que recortamos el valor dentro del rango usando la funcion clip() de Numpy
'''


theta = vector_angle(u, v)
print("Angle =", theta, "radians")
print("      =", theta * 180/ np.pi, "degrees")
```

#### Proyectar un punto sobre un eje

![Proyeccion de vectores](./images/Vectores-parte10.jpg)

```python
u_normalized = u / LA.norm(u)
proj = v.dot(u_normalized) * u_normalized

plot_vector2d(u, color="r")
plot_vector2d(v, color="b")

plot_vector2d(proj, color="k", linestyle=":")
plt.plot(proj[0], proj[1], "ko")

plt.plot([proj[0], v[0]], [proj[1], v[1]], "b:") # linea azul punteada entre v y la proyeccion de v sobre u

plt.text(1, 2, "$proj_u v$", color="k", fontsize=18)
plt.text(1.8, 0.2, "$v$", color="b", fontsize=18)
plt.text(0.8, 3, "$u$", color="r", fontsize=18)

plt.axis([0, 8, 0, 5.5])
plt.grid()
plt.show()
```

### Ejemplo: Embeddings

<p style="font-size:1rem; text-align: justify">Los Embeddings son un caso de uso de vectores en ML donde dichos vectores se aprovechan para representar algo (un objeto, una persona, etc) mediante los componentes numéricos del vector, donde cada uno de estos componentes hace alusión a alguna característica en forma de medida.</p>

#### Embedding de personalidad: ¿Cómo somos?

<p style="font-size:1rem; text-align: justify">En una escala de 0 a 100, ¿qué tan introvertido/extrovertido eres? (0 es más introvertido y 100 es el más extrovertido)
    <br>
Un ejemplo de test de personalidad para saberlo es la prueba de los Cinco Grandes Rasgos de Personalidad, donde te realizan una lista de preguntas y te califican en una serie de ejes, siendo el de la introversión/extraversión uno de dichos ejes.
</p>

![Embedding](./images/Vectores-parte11.jpg)

Pero, si debemos tener en cuenta u vector con los 5 rasgos para cada persona, el análisis gráfico ya no es una opción.<br>
Sin embargo, podemos calcular la siilaridad a través del ángulo entre vectores (por mencionar una de las operaciones entre vectores vistas), el cual se calcula utilziando el coseno, para obtener una métrica que facilite la comparación.<br>
Eso implicaría que cuanto menor sea el ángulo, están aún más cerca de apuntar en la misma dirección y sentido y, por lo tanto, de ser parecidos. <br>

```python
x = np.array([180, 157, 162, 167, 178, 178, 182, 188]) # mi vector x = [x1, x2, ..., xn]
y = np.array([56, 61, 67, 72, 70, 72, 83, 92]) # mi vector y = [y1, y2, ..., yn]
f, ax = plt.subplots(1)
ax.scatter(x, y)
ax.set_xlim(xmin=140, xmax=200)
ax.set_ylim(ymin=40, ymax=100)
plt.show(f)
```

<p style="font-size:1rem; align: justify">Programaremos las fórmulas de regresión lineal en base a la teoría vista y graficaremos la recta encontrada:</p>

```python
'''
Aqui tiene lugar el bradcasting. La funcion mean() nos dara el promedio de los componentes del vector, 
pero justamente nos devolvera un escalar y luego estamos queriendo restarle al vector x dicho escalar.
Numpy considera que lo que intentamos hacer en realidad es restar a cada componente del vector x dicho escalar,
y eso se llama broadcasting.
En definitiva, el bradcasting se activa siempre que operamos un vector con un escalar
'''
b1_numerador = (x - x.mean()).dot(y - y.mean())
b1_denominador = (x - x.mean()).dot(x - x.mean())
b1 = b1_numerador / b1_denominador
b0 = y.mean() - b1*x.mean()
print(b1)
print(b0)
y_hat = b0 + b1*x # Nuestra predicción (la y de los puntos x de la recta, o sea el vector y sombrero)

# Graficamos
f, ax = plt.subplots(1)
ax.scatter(x, y) # Nube de puntos
ax.plot(x, y_hat, "go--", linewidth=2) # Recta de nuestro modelo/prediccion
ax.set_xlim(xmin=140, xmax=200)
ax.set_ylim(ymin=40, ymax=100)
plt.show(f)
```

<p style="font-size:1rem; align: justify">Ahora veremos el valor del error:</p>

```python
e_vec = y - y_hat
e = np.dot(e_vec, e_vec)
e # tener en cuenta que este no seria el valor directo para entender el error porque se trata del cuadrados sumados
```
