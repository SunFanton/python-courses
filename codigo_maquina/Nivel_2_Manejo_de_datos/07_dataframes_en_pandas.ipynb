# Creación y Manipulación de un DataFrame

## Creación de un DataFrame

```python
import pandas as pd
```

Los DataFrames con conjuntos de Series. Debemos pensarlos como "tablas", que tiene un conjunto de columnas (que son las series). Para el análisis usamos 10 países. De cada uno tenemos la población (en millones) y el porcetaje que representa esa población en el total mundial.

```python
nombre_paises = ["China", "India", "Estados Unidos", "Indonesia", "Pakistán",
                 "Brasil", "Nigeria", "Bangladesh", "Rusia", "México"]

encabezado = ["poblacion", "porcentaje"]

datos = [[1439, 18.47],
        [1380, 17.70],
        [331, 4.25],
        [273, 3.51], 
        [220, 2.83],
        [212, 2.73], 
        [206, 2.64],
        [164, 2.11],
        [145, 1.87],
        [128, 1.65]]

# nos crea un dataframe con indices y encabezados numericos por defecto, empezando en cero
paises = pd.DataFrame(datos) 
paises
```

```python
# podemos pasar los parametros para los indices (de las filas) y encabezados de cada columna
paises = pd.DataFrame(datos, index=nombre_paises, columns=encabezado)
paises
```

```python
# otra forma de crear dataframes pero a partir de un diccionario:
datos = {"China": [1439, 18.47],
         "India": [1380, 17.70],
         "Estados Unidos": [331, 4.25],
         "Indonesia": [273, 3.51], 
         "Pakistán": [220, 2.83],
         "Brasil": [212, 2.73], 
         "Nigeria": [206, 2.64],
         "Bangladesh": [164, 2.11],
         "Rusia": [145, 1.87],
         "México": [128, 1.65]}

paises = pd.DataFrame(datos, index = encabezado)
paises = paises.transpose() # equivalente a paises.T
paises
```

## Agregar y borrar renglones y columnas

```python
paises
```

```python
tasa_fertilidad = [1.7, 2.2, 1.8, 2.3, 3.6, 1.7, 5.4, 2.1, 1.8, 2.1]
paises["tasa_fertilidad"] = tasa_fertilidad
paises
```

```python
# eliminar columna (pop devuelve la columna eliminada)
paises.pop("tasa_fertilidad")
```

```python
paises
```

```python
del paises["tasa_fertilidad"]
```

```python
paises
```

```python
# drop puede borrar columnas y renglones
paises.drop("tasa_fertilidad", axis=1, inplace=True)
```

```python
paises
```

```python
# Japón -> 126, 1.62
renglon = pd.Series(name="Japón", data=[126, 1.62], index=["poblacion", "porcentaje"])
renglon
```

```python
paises = pd.concat([paises, renglon.to_frame().T])
```

```python
paises
```

```python
paises.drop(["Bangladesh", "Japón"], axis=0, inplace=True)
```

```python
paises
```
