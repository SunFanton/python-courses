### Step 8. Group the year by decades and sum the values

#### Pay attention to the Population column number, summing this column is a mistake

En este caso se puede utilizar el método resample() de Pandas.

El atributo *rule* de dicho método permite determinar la fercuencia de días que se desea tomar.

El parámetro rule de la función resample() en pandas acepta una variedad de cadenas que especifican la frecuencia de muestreo. Aquí tienes un listado de algunas de las opciones más comunes:
Frecuencias de tiempo:

    A: Año (fin de año)
    Q: Trimestre (fin de trimestre)
    M: Mes (fin de mes)
    W: Semana (fin de semana, domingo por defecto)
    D: Día
    B: Día laboral (excluye fines de semana)
    H: Hora
    T o min: Minuto
    S: Segundo
    L o ms: Milisegundo
    U o us: Microsegundo
    N: Nanosegundo

Frecuencias personalizadas:

Puedes combinar las letras para crear frecuencias personalizadas. Algunos ejemplos incluyen:

    5D: Cada 5 días
    3M: Cada 3 meses
    2H: Cada 2 horas

Frecuencias de negocio:

    BM: Último día laboral del mes
    BQ: Último día laboral del trimestre

Otras opciones:

    S: Segundo
    T: Minuto

```python
crimes_by_decades = crime.resample(rule="10AS").sum()
crimes_by_decades["Population"]= crime["Population"].resample(rule="10AS").max()
crimes_by_decades.set_index(keys=crimes_by_decades.index.year, inplace=True)
crimes_by_decades
```

### Step 9. What is the most dangerous decade to live in the US?

```python
# Forma 1
crime.idxmax(0)
```

```python
# Forma 2
# Aqui la idea es obtener un total general y compara ese total por decada con la cantidad de poblacion

crimes_by_decades["Total"] = crimes_by_decades.drop(columns="Population").sum(axis=1)
crimes_by_decades["Ratio"] = crimes_by_decades["Total"] / crimes_by_decades["Population"]
crimes_by_decades.head()
```

```python
crimes_by_decades.nlargest(2, "Ratio")
```
