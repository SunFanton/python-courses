# Correlación de Pearson

La correlación de Pearson se utiliza para indicar el grado o el coeficiente de relación entre dos variables. 
Es decir, es una forma de tener una noción de qué sucede cuando el valor de una de las variables a analizar aumenta, qué ocurre con la otra. Esta segunda variable aumenta también? Disminuye? O es indstinto?

Para introducir el concepto de una manera más formal, acudí a material que aporta una base desde el punto de vista matemático y de las fórmulas que nos llevan al cálculo de esta correlación. Y luego se abordará la misma mediante el uso de Python y las funciones y métodos que ya se encuentran predefinidos para la misma.

## Coeficiente de correlación de Pearson

### Concepto y fórmula

El coeficiente de correlación de Pearson viene a brindarnos una ampliación de la información que nos da la covarianza, de una manera más clara y precisa para entender cómo se relacionan los valores de dos variables.

Para entender el concepto, me he basado en los siguientes videos de Youtube:

* [Coeficiente de correlación de Pearson](https://www.youtube.com/watch?v=eTuGHN53kJQ)
* [Ejemplo correlación de Pearson](https://www.youtube.com/watch?v=1BMX762A9Dg)

Es importante tener en cuenta las siguientes afirmaciones, entendiendo que la correlación NO implica necesariamente causalidad, sino que expresa cómo se relacionan dos variables entre sí:

* Cuando entre las dos variables hay una **relación directa**, la covarianza da un **valor positivo**.
* Cuando entre las dos variables hay una **relación inversa**, la covarianza da un **valor negativo**.
* Cuando entre las dos variables **no hay una relación**, la covarianza da un **valor en torno a cero**.

En el primer caso, lo que implica es que cuando una variable aumenta en su valor, se observa que la otra variable también aumenta, de allí que se diga que su relación es directa. Gráficamente es así:

![Relación directa](./img/pearson1.png)

En el segundo caso, cuando una variable aumenta su valor, la otra disminuye, de allí la relación inversa entre ambas. Gráficamente:

![Relación directa](./img/pearson2.png)

Por último, en el tercer caso, si una variable aumenta su valor, será indistinto para la otra variable, justamente porque no hay una relación entre ellas. Gráficamente:

![Relación directa](./img/pearson3.png)

La fórmula del coeficiente de correlación de Pearson es la siguiente:

![Relación directa](./img/pearson4.png)

Por otro lado, debemos tener en cuenta una serie de características importantes de dicho coeficiente:

* $r$ es un coeficiente (no tiene unidades, ej cm, lts, etc)
* $r$ se encuentra en el siguiente intervalo: $-1 \leq r \leq 1$

  * Si $r \approx 1$ existe una correlación directa fuerte
  * Si $r \approx -1$ existe una correlación inversa fuerte
  * Si $r = 1$ o $r = -1$ hay una correlación funcional (es decir, la relación entre ambas variables está dada por una función, ej función lineal)
  * Si $r \approx 0$ no existe una correlación *lineal* (pero pueden estar relacionadas por otro tipo de funciones)

### Correlación de Pearson en Python

#### Magnitud y dirección de una relación lineal entre dos variables

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
personas = pd.read_csv("../datos/personas.csv")
personas.head()
```

```python
personas.shape[0]
```

##### Relación altura y peso

```python
plt.scatter(personas["altura"], personas["peso"])
plt.xlabel("Altura (cms)")
plt.ylabel("Peso (kgs)")
plt.title("Relación lineal directa entre altura y peso")
plt.show()
```

##### Relación Ingreso y Horas trabajadas

```python
plt.scatter(personas["ingreso"], personas["horas_trabajadas"])
plt.xlabel("Ingreso ($)")
plt.ylabel("Horas trabajadas")
plt.title("Relación entre ingreso y horas trabajadas")
plt.show()
```

##### Relación Ingreso y Ausencias

```python
plt.scatter(personas["ingreso"], personas["ausencias"])
plt.xlabel("Ingreso ($)")
plt.ylabel("Ausencias")
plt.title("Relación lineal inversa entre ingreso y ausencias")
plt.show()
```

##### Relación Peso e Ingreso

```python
plt.scatter(personas["peso"], personas["ingreso"])
plt.xlabel("Peso (kgs)")
plt.ylabel("Ingreso ($)")
plt.title("Relación entre peso e ingreso")
plt.show()
```

#### Matriz de Correlación

El método `corr()` nos devuelve un dataframe, es decir, una matriz que es la matriz de correlación. La misma posee los datos con los coeficientes de correlación entre todos los pares posibles del dataframe con nuestros datos. Por defecto, aplica el método de correlación de Pearson.

```python
matriz = personas.corr()
matriz
```

No existe una referencia exacta para saber qué valor se considera una correlación alta (directa o inversa) o baja.
Por lo general, existe esta tendencia (siendo $r$ el coeficiente de correlación):

* $-0.5 \leq r \leq -0.3$ o bien $0.3 \leq r \leq 0.5$: se considera una correlación baja
* $-0.7 \leq r \leq -0.5$ o bien $0.5 \leq r \leq 0.7$: se considera una correlación moderada
* $-1 \leq r \leq -0.7$ o bien $0.7 \leq r \leq 1$: se considera una correlación fuerte

De todas formas, esto depende del contexto de los datos y del tamaño de la muestra.

#### Mapa de calor

```python
plt.matshow(matriz, cmap="bwr", vmin=-1, vmax=1) # mapa de calor con: blue(-1) white(0) red(1
plt.xticks(range(5), personas.columns, rotation=90)
plt.yticks(range(5), personas.columns)
plt.colorbar()
plt.title("Mapa de calor de matriz de correlación Personas")

for i in range(len(matriz.columns)):
    for j in range(len(matriz.columns)):
        plt.text(i, j, round(matriz.iloc[i, j], 2), ha="center", va="center")

plt.show()
```

```python
# usando seaborn podemos graficar el mapa de calor aún más fácilmente
sns.heatmap(matriz, annot=True)
plt.title("Mapa de calor de matriz de correlación Personas")
plt.show()
```
