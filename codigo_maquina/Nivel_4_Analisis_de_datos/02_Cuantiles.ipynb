# Cuantiles

Los cuantiles son valores que nos permiten particionar los datos en intervalos idénticos. Los diferentes cuantiles se dan en función de cuántas particiones se realizan de la distribución de los datos.

![cuantiles](./img/cuantiles1.png)

En el ejemplo que se verá a continuación, se trabajará con la altura de un grupo de personas y dichos datos tendrán una distribución normal.

En particular, el percentil nro 100 representará el valor máximo de nuestro datos, es decir, la mayor altura. Mientras que el percentil 0 representará el menor valor, o sea la menor altura.
El percentil 10 es equivalente al decil 1 (los deciles dividen en 10 partes los datos), lo que significa que el 10% de las personas de nuestra muestra tendrán una altura menor o igual al valor de dicho percentil/decil.
El percentil 20 es equivalente al decil 2, es decir, que el 20% de las personas de nuestra muestra tienen una altura menor o igual al valor de dicho percentil/decil y así sucesivamente.

En el caso del percentil 50, será equivalente al decil 5 pero en el caso de la distribución normal también será equivalente a la mediana.

## Cálculo de Cuantiles con Numpy

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
alturas = np.random.normal(1.75, 0.15, 100000) # media, desvio, tamaño muestra
alturas
```

```python
# el vector representa que quiero el minimo (0), el cuartil 1, el cuartil 2, el cuartil 3 y el maximo
np.quantile(alturas, [0, 1/4, 2/4, 3/4, 1])
```

```python
# terciles
np.quantile(alturas, [0, 1/3, 2/3, 1])
```

```python
cuantiles = [
    (2, "Mediana"),
    (3, "Terciles"),
    (4, "Cuartiles"),
    (5, "Quintiles"),
    (10, "Deciles"),
    (100, "Percentiles")
]


cortes = []

for cuantil in cuantiles:
  corte = []
  valor = 0
  for i in range(cuantil[0] - 1):
    valor += 1/cuantil[0]
    corte.append(valor)
  cortes.append(corte)

# print(cortes)

for i in range(len(cuantiles)):
  plt.figure(figsize=(7.5, 5))
  plt.title(cuantiles[i][1], size=22)
  plt.ylabel("Frecuencia", size=16)
  plt.xlabel("Alturas", size=16)
  plt.hist(alturas, 100, label="altura", color="pink")
  #print(np.quantile(alturas, [0] + cortes[i] + [1]))
  for corte in cortes[i]:
      plt.axvline(x = np.quantile(alturas, corte), label="%.2f" % corte, linewidth=3) # linea vertical
  if i < len(cuantiles) - 1:
      plt.legend()
  plt.show()
  print("\n"*4)
```
