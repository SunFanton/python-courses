# Creación de Diagramas de Caja (BoxPlots)

```python
import matplotlib.pyplot as plt
import numpy as np
```

El gráfico **boxplot** consiste en un diagrama tipo "caja" que nos muestra información relevante de nuestros datos, los cuales deben encontrarse ordenados. Las líneas superior e inferior nos muestran el valor máxmo y mínimo respectivamente. La línea del medio (en este caso, la línea naranja) nos muestra la mediana, es decir, el valor que se encuentra en el medio de nuestro conjunto de datos (no confundir con la media o promedio). Y por otro lado los cuartiles. El primer cuartil es la línea inferior de la caja. El segundo sería la línea naranja. El tercer cuartil se encuentra en el límite superior de la caja.

![Cuartiles](./cuartiles.png "Cuartiles")

```python
edades = np.array([4, 17, 29, 33, 50, 51, 61, 71, 85, 92, 99])

plt.boxplot(edades)
plt.yticks(np.arange(0, 110, 5))
plt.grid()
plt.show()
```

```python
(edades.min(), np.quantile(edades, 0.25), np.median(edades), np.quantile(edades, 0.75), edades.max())
```

# Datos Anormales / Atípicos / Outliers

## Prueba de Tukey

Puede darse el canso de que entre nuestros datos se encuentren datos anómalos, atípicos o outliers. Estos son aquellos que no se corresponden con la mayoría de los valores de nuestros datos. En el ejemplo de abajo, vemos que tenemos un grupo de edades que rondan entre los 40 y 65 años, y después otros tres valores que se encuentran muy alejados de ese rango, como el 6, 95 y 100. Estos últimos serían los valores atípicos. En el contexto de análisis de datos y machine learning, si no se tienen en cuenta dichos datos, podemos encontrarnos con análisis o entrenamientos sesgados e incorrectos. 

Una forma de detectarlos es mediante la prueba de Tukey que se basa por un lado en las reglas empíricas de la distribución normal y en el rango intercuartílico para poder hallar dichos outliers. Eso lo vemos claramente reflejado en el boxplot, donde los "círculos" hacen referencias a los mismos. 

![Distribución Normal](./distribucion_normal_regla_empirica.png "Distribución Normal")

![Rango Intercuartilico](./rango_intercuartilico.png "Rango Intercuartilico")

```python
edades = np.array([6, 
                   41, 51, 51, 52, 54, 57, 60, 61, 65,
                   95, 100])

plt.boxplot(edades)
plt.yticks(np.arange(0, 110, 5))
plt.grid()
plt.show()
```
