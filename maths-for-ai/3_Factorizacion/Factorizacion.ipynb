# Factorización

## ¿Qué es la factorización de matrices?

<p style="text-align: justify; font-size:1.1rem;">Es la expresión de una matriz en término del producto matricial de otras matrices (dos o más matrices). La idea es que la expresión de ese producto entre matrices se simplifique mediante una sola matriz. De esa forma, cada elemento o componente de la matriz ${M}$ podrá expresarse como la suma de los productos entre los componentes/coeficientes de las otras matrices. Por ejemplo:</p>

$M = A.B.C$

$$\begin{bmatrix} M_{1,1} & M_{1,2} \\ M_{2,1} & M_{2,2} \end{bmatrix} = \begin{bmatrix} A_{1,1} & A_{1,2} \\ A_{2,1} & A_{2,2} \end{bmatrix}. \begin{bmatrix} B_{1,1} & B_{1,2} \\ B_{2,1} & B_{2,2} \end{bmatrix}.\begin{bmatrix} C_{1,1} & C_{1,2} \\ C_{2,1} & C_{2,2} \end{bmatrix} $$

<p style="text-align: justify; font-size:1.1rem;">La importancia de las matrices radica, entre otras cosas, en la simplificación para la aplicación de <b>transformaciones lineales</b> sobre un <b>espacio vectorial</b>. Es fundamenteal tener en cuenta que las transformaciones lineales viven y usan los elementos que conforman un espacio vectorial.</p>

## ¿Qué es un espacio vectorial?

<p style="text-align: justify; font-size:1.1rem;">Es un conjunto V de elementos, dotado de dos operaciones características:</p>
<p style="text-align: justify; font-size:1.1rem;">
* La suma de elementos de V <br>
* La multiplicación de un elemento de V por un escalar</p>

<p style="text-align: justify; font-size:1.1rem;">Ambas operaciones tienen a su vez la característica de ser <i>cerradas</i>, es decir, si yo tomo dos elementos cualquiera del espacio y los sumo, o tomo un elemento y lo multiplico por un escalar, en ambos casos voy a obtener como resultado otro elemento del mismo espacio vectorial</p> 

<p style="text-align: justify; font-size:1.1rem;">Por otra parte, un conjunto de elementos se considera un espacio vectorial siempre y cuando se cumplan las siguientes propiedades: </p>

<div style="text-align: justify; font-size:1.1rem;">

1. La suma es conmutativa: <br>
    
    $u + v = v + u$ <br>

2. La suma es asociativa: <br>

   $(u + v) + w = u + (v + w)$ <br>

4. Existe un único vector nulo $0$ (por ejemplo, para vectores de 2 dimensiones, sería el vector (0, 0) ) tal que <br>

    $v + 0 = v$ <br>

5. Existencia del opuesto: Existe  $-v$ tal que <br>

    $v + (-v) = 0$<br>

6. Multiplicación por un escalar distributiva <br>

    $\alpha(u + v) = \alpha u + \alpha v$ <br>

7. Suma de escalares distributiva respecto del producto por vectores <br>

    $(\alpha + \beta)v = \alpha v + \beta v$ <br>

8. Multiplicación por escalares distributiva <br>

    $\beta(\alpha v) = (\beta\alpha) v$ <br>

9. Existencia del elemento neutro de la multiplicación de vectores por escalares <br>

    $1v = v$
</div>

## ¿Qué es una transformación lineal?

<p style="text-align: justify; font-size:1.1rem;">Una función $L : V \Rightarrow  W$, que toma elementos de un espacio vectorial, los transforma y da como resultado elementos de otro espacio vectorial o incluso del mismo espacio inicial. Va de un espacio vectorial $V$ en otro espacio vectorial $W$ (ambos sobre el mismo conjunto de escalares) y es una transformación lineal si satisface:</p>

<div style="text-align: justify; font-size:1.1rem;">

$L (\alpha v_1 + \beta v_2) = \alpha L (v1) + \beta L (v2)$ <br>

para todo par de vectores $v_1$ y $v_2$ de $V$ y todo par de escalares $\alpha$, $\beta$. <br>

Es decir, esto puede sintetizarse en dos propiedades: <br>
$L (v_1 + v_2) = L (v_1) + L (v_2)$ <br>

$L (\alpha v) = \alpha L (v) $ <br>
</div>

### Transformaciones lineales representadas como matrices

<p style="text-align: justify; font-size:1.1rem;">Es posible representar y operacionalizar las transformaciones lineales sobre el espacio vectorial de vectores de n dimensiones con matrices y vectores de dimensiones apropiadas. Es decir, yo puedo representar una transformación lineal mediante una matriz y la forma de aplicar la transformación lineal es multiplicando esa matriz (lado izquierdo de la multiplicación) por un vector del espacio vectorial inicial ${V}$ (vector que se ubicará del lado derecho de la multiplicación).</p>

<p style="text-align: justify; font-size:1.1rem;">En el ejemplo genérico de abajo, si bien estoy utilizando una matriz de 2x2 y un vector de 2 filas (para que se pueda definir la multiplicación), tener en cuenta que esta noción de representación de una TL (transformación lineal) a través de una matriz puede generalizarse a matrices más grandes también.</p>

<div style="text-align: justify; font-size:1.1rem;">
    
$L (v) = M.v =\begin{bmatrix} M_{(1,1)} & M_{(1,2)} \\ M_{(2,1)} & M_{(2,2)} \end{bmatrix}.\begin{bmatrix} v_{1} \\ v_{2}  \end{bmatrix} = \begin{bmatrix} M_{(1,1)}v_1 + M_{(1,2)}v_2 \\ M_{(2,1)}v_1 + M_{(2,2)}v_2 \end{bmatrix} $

</div>


<p style="text-align: justify; font-size:1.1rem;">La transformación lineal $L$ es representada a través de la matriz $M$</p>

<p style="text-align: justify; font-size:1.1rem;">Las operaciones matriciales (suma y multiplicación de matrices), las operaciones entre escalares (suma y multiplicación de escalares) y las operaciones entre matrices y escalares (multiplicación de una matriz por un escalar) permiten que las condiciones del espacio vectorial y las transfomraciones lineales se cumplan en dicha representación.</p>

<p style="text-align: justify; font-size:1.1rem;">La "linealidad" de las transfomraciones lineales implican que las componenetes resultantes de los vectores sean sumas de las componentes originales (elevadas a la 1) multiplicadas por algun escalar y no contempla efectos no lineales (potencias distintas de 1 u otras funciones). Es decir, no voy a obtener vectores que sean con componentes elevados a la 2 o más, ni otras funciones como puede ser logaritmos, etc, que no estén relacionadas con la linealidad.</p>

### Transformaciones lineales aplicadas a un vector

<p style="text-align: justify; font-size:1.1rem;">Una transformación lineal aplicada a un vector, va a devolverme como resultado otro vector:</p>

<div style="text-align: justify; font-size:1.1rem;"> 

$L (v_1) = v_2$ 

</div>

<p style="text-align: justify; font-size:1.1rem;">Por ejemplo:</p>

````python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon
````

````python
def plot_vector2d(vector2d, origin=[0,0], **options):
    return plt.arrow(origin[0], origin[1], vector2d[0], vector2d[1],
                     head_width=0.2, head_length=0.3, length_includes_head=True,
                     **options)
````

````python
v1 = np.array([2, 1]) # defino un vector cualquiera al que le desee aplicar una TL
v1
````

````python
# defino una matriz que utilizare para aplciar una TL
M = np.array([
    [0.5, 0.5],
    [1, 1]
]) 
# v2 = L(v1)
v2 = M @ v1 # Aplico la transformación lineal mediante el producto matricial entre la matriz M y el vector v1 (igual que M.dot(v) )

plot_vector2d(v1, color="blue")
plot_vector2d(v2, color="red")
plt.text(1.5, 0.5, "${v1}$", fontsize=18, color="blue")
plt.text(0.80, 2.5, "${v2}$", fontsize=18, color="red")
plt.ylim(0,3.5)
plt.xlim(0,3)
plt.grid()
plt.show()
````

### Transformaciones lineales aplicadas a un conjunto de vectores

<p style="text-align: justify; font-size:1.1rem;">Una transformación lineal aplicada a un conjunto de vectores me va a devolver otro conjunto de vectores. Si bien en el ejemplo anterior pudimos ver cómo una matriz que representa una transformación lineal multiplicada por un vector nos dio como resultado un nuevo vector, nos quedaría más claro ver cómo actúa esta matriz y TL sobre un conjunto de vectores para tener una idea más general del tipo de transformación que se lleva a cabo.</p>

<p style="text-align: justify; font-size:1.1rem;">Dicho de otro modo, es útil representar el resultado de una transformación lineal sobre un conjunto de vectores que conforman una determinada figura geométrica para visualizar de manera más sencilla sus efectos.</p>

````python
def mark_all_arrows(vector, **options):
    rows, _ = vector.shape
    colors = ["cyan", "magenta", "yellow", "blue"]
    for row in range(0, rows):
        v = np.array(vector[row][:]).flatten()
        plot_vector2d(v, color=colors[row])


# Funcion que dibuja una figura (en este caso, un cuadro centrado en el origen)

def initial():

  colors = ["c", "m", "y", "b"] # cyan, magenta, yellow, black

  X = [[2*i-1,2*j-1] for i in range(2) for j in range(2)] 
  # me da como resultado: 
  """ [
            [-1,-1], 
            [-1, 1], 
            [1, -1], 
            [1, 1]
        ] """

  # Se cambia el orden de la fila 4 y 3
  aux = X[2][:]
  X[2][:] = X[3][:]
  X[3][:] = aux

  # Se traspone la matriz X, quedando una matriz de 2 x 4
  X = np.matrix(X).T # np.matrix(X).transpose()

  # Se agrega una figura de 4 x
  plt.figure(figsize=(4, 4))
  ax1 = plt.subplot(111)
  ax1.scatter(X[0,:].flat, X[1,:].flat, color = colors,
              s = 12*16, alpha = 1.0, marker = "s")   
  ax1.scatter((2 * X[0,:]).flat, (2* X[1,:]).flat, color = colors,
              s = 12*16, alpha = 0.0, marker = "s")
  ax1.add_patch(Polygon(xy = X.T, color = "g", alpha = 1)) 
  mark_all_arrows(X.T, color="black")
  plt.grid()
  plt.show()



# Funcion que aplica la TL a la figura anterior mediante la matriz pasada por parametro y dibuja el resultado

def show_TL(T): 
  T = np.matrix(T)
  if T.shape != np.shape(np.eye(2)):
    print("[INFO] bad matrix! it's not squared")
    return
  X = [[2*i-1,2*j-1] for i in range(2) for j in range(2)] # Generamos la mariz cuadrada de longitud 2 de cada lado
  aux = X[2][:]
  X[2][:] = X[3][:]
  X[3][:] = aux
  X = np.matrix(X).T # np.matrix(X).transpose()
  TX = T @ X # Aplica la TL multiplicando T por X

  w = TX[0,:].max() - TX[0,:].min()
  h = TX[1,:].max() - TX[1,:].min()

  if w < 2:
    w = 2
  if h < 2:
    h = 2
  while w > 5 or w > 5:
    h *= 0.99
    w *= 0.99

  colors = ["c", "m", "y", "b"]
  

  plt.figure(figsize=(3*w*1.2, h*1.2))

  ax1 = plt.subplot(131)
  ax1.set_axis_off()
  ax1.add_patch(Polygon(xy = X.T, color = "g", alpha = 1.0))
  ax1.add_patch(Polygon(xy = TX.T, color = "k", alpha = 0.0))
  ax1.scatter(X[0,:].flat, X[1,:].flat, color = colors,
              s = 12*h*w, alpha = 1.0, marker = "s")  
  ax1.scatter(TX[0,:].flat, TX[1,:].flat, color = colors,
              s = 12*h*w, alpha = 0.0, marker = "s")  

  ax2 = plt.subplot(132)
  ax2.set_axis_off()
  ax2.add_patch(Polygon(xy = X.T, color = "g", alpha = 0.0))
  ax2.add_patch(Polygon(xy = TX.T, color = "k", alpha = 1.0))
  ax2.scatter(X[0,:].flat, X[1,:].flat, color = colors,
              s = 12*h*w, alpha = 0.0, marker = "s")  
  ax2.scatter(TX[0,:].flat, TX[1,:].flat, color = colors,
              s = 12*h*w, alpha = 1.0, marker = "s")  

  ax3 = plt.subplot(133)
  ax3.add_patch(Polygon(xy = X.T, color = "g", alpha = 0.5))
  ax3.add_patch(Polygon(xy = TX.T, color = "k", alpha = 0.5))
  ax3.scatter(X[0,:].flat, X[1,:].flat, color = colors,
              s = 12*h*w, alpha = 0.0, marker = "s")  
  ax3.scatter(TX[0,:].flat, TX[1,:].flat, color = colors,
              s = 12*h*w, alpha = 0.0, marker = "s")  

  mark_all_arrows(X.T, color="black")
  mark_all_arrows(TX.T, color="blue")
  plt.show()
````

````python
M = [[1,0.5],[0.5,2]]
show_TL(M)
````

<p style="text-align: justify; font-size:1.1rem;">Más allá de responder a estas preguntas, para el ejemplo que planteamos abajo, hemos de entender algunas cuestiones.</p>

<p style="text-align: justify; font-size:1.1rem;">Es posible aplicar varias transformaciones lineales a un vector a través de varias matrices. Pero para no hacerlo de a una matriz a la vez, es decir, multiplicar el vector por la matriz de la primera TL, y al resultado por la segunda matriz y así sucesivamente, se puede aplicar el concepto de encadenamiento de matrices. Es decir, en primer lugar multiplico las matrices de las TL que quiero aplicar entre sí, lo cual almaceno en una matriz y es esta matriz la que luego multiplico por el vector en cuestión que deseo transformar.</p>

<p style="text-align: justify; font-size:1.1rem;">Si tomamos el ejemplo de abajo (matriz L), L es una matriz que surge de este encadenamiento de matrices, cada una de las cuales representa una TL. Tener en cuenta en especial que el orden en que se encadenan estas matrices influye en el orden en que se aplica cada TL. En el caso de L, la última matriz del encadenamiento es R.T (R transpuesta que resulta ser igual a R inversa en este caso particular). Como está al final, esa sería la primer transformación lineal que se aplicaría en el vector, que sería una rotación. A continuación, se aplicaría la transformación lineal de F (matriz de reflexión), que es la que le sigue a la izquierda a R.T. Y finalmente, la que le sigue a F es R, con lo cual la útlima transformación aplicada sería de nuevo la misma rotación, pero en sentido contrario.</p>

<p style="text-align: justify; font-size:1.1rem;">El encadenamiento de matrices para representar un conjunto de TLs siempre actúa aplicando cada transformación de <i>derecha a izquierda</i>. Es una forma de combinar distintas TLs para generar transformaciones consecutivas y más complejas.</p>

![Ejemplo de TL](./images/Factorizacion-parte8.jpg)

````python
# Creamos la matriz K con elementos no nulos e iguales en la diagonal principal y ceros en el resto
K = [[2,0],[0,2]]

# Para RECORDAR:
# Matriz de reflexion (sobre y)
#F = [[-1,0],
#     [0,1]]

# Matriz de rotacion:
#R = [[3/5, -4/5],
#      4/5,  3/5]]


"""
En el caso de la matriz de rotacion R utilizada en los ejemplos, la matriz inversa es igual a la matriz transpuesta. De alli que abajo 
estemos utilizando R.T en lugar de buscar la inversa con la funcion de Numpy np.inv(...)
"""

L = np.matrix(F)
L = R @ F @ R.T 

# Matriz de proyeccion sobre un eje (en este caso, sobre eje y, anulando valores de x)
#PI = [[0,0],
#      [0,1]]

P = np.matrix(PI)
P = R @ P @ R.T

print("P")
print(P)
print("L")
print(L)

# Respuesta a la pregunta: ¿Podemos representar la multiplicación de un vector por un escalar con una matriz? Sí
"""
En este caso, K representa una matriz que duplica el tamaño de los vectores (si en vez de 2 hubiera sido un numero menor a 1, lo hubiera disminuido).
Y esto es tal cual sucede como cuando multiplicamos un vector por un escalar, el cual, según su valor, aumenta o reduce la longitud
del vector en cuestion
"""
show_TL(K)

# Este es para tener una referencia de la rotacion que genera la matriz R.T
show_TL(R.T)

# Respuesta a la pregunta: ¿Podemos representar la proyección en un eje arbitrario con una matriz? Sí
"""
Notar que aqui se estan aplicando las TLs encadenadas que se agrupan en la matriz P.
Primero se gira la figura, luego se comprime o proyecta en el eje y (todas las x pasan a ser cero), y luego se vuelve a rotar, pero 
en sentido contrario.
Esto da la impresion de que se proyecta sobre un eje arbitrario (un eje distinto de los tipicos x e y)
"""
show_TL(P)

# Respuesta a la pregunta: ¿Podemos representar la reflexión por un eje arbitrario con una matriz? Sí
"""
Notar que aqui se estan aplicando las TLs encadenadas que se agrupan en la matriz L.
Primero se gira la figura, luego se refleja (usando el eje y), y luego se vuelve a rotar, pero 
en sentido contrario.
Esto da la impresion de que se refleja sobre un eje arbitrario (un eje distinto de los tipicos x e y)
"""
show_TL(L)
````

### Limitaciones de una transformación lineal

<div style="text-align: justify; font-size:1.1rem;">
Nos quedó responder a estas dos preguntas:

* ¿Podemos representar el cálculo del determinante con una matriz?
* ¿Podemos representar el cálculo de la inversa con una matriz?
    
Hemos visto que hay cosas que una matriz o una transformación lineal no puede hacer. Sería interesante saber cuales son los límites de una transformación lineal.

Por definición, las transformaciones lineales deben preservar sumas y multiplicaciones por un escalar.

Si $M$ es una transformación lineal, $x$ e $y$ vectores y α un escalar entonces M debe verificar:

$M(x + y) = Mx + My$

$M(\alpha x) = \alpha(Mx)$

Para todo $x$, $y$, $\alpha$

A partir de la definición anterior, se puede ver que la inversa y el determinante no pueden ser transformaciones lineales
</div>

#### Inversa de una matriz

<div style="text-align: justify; font-size:1.1rem;">

$\mathrm{Inv}(I) = 
\mathrm{Inv}\left(\begin{bmatrix}
1 & 0\\
0 & 1\\
\end{bmatrix}\right)=\begin{bmatrix}
1 & 0\\
0 & 1\\
\end{bmatrix}$

$\mathrm{Inv}(0I) = 
\mathrm{Inv}\left(0\begin{bmatrix}
1 & 0\\
0 & 1\\
\end{bmatrix}\right)=0\mathrm{Inv}\left(\begin{bmatrix}
1 & 0\\
0 & 1\\
\end{bmatrix}\right)=0\begin{bmatrix}
1 & 0\\
0 & 1\\
\end{bmatrix}=\begin{bmatrix}
0 & 0\\
0 & 0\\
\end{bmatrix}$

¡Pero la matriz $0I$ No tiene inversa!
</div>

#### Determinante de una matriz

<div style="text-align: justify; font-size:1.1rem;">

$\det(I)=\det\left(\begin{bmatrix}
1 & 0\\
0 & 1\\
\end{bmatrix}\right) = 1$

$\det(2I) = \det(I + I) = \det(I) + \det(I) = 1 + 1 = 2$

Pero:

$\det(2I)=\det\left(\begin{bmatrix}
2 & 0\\
0 & 2\\
\end{bmatrix}\right) = 4$

Entonces, no es lo mismo $\det(2I)$ que $2\det(I)$
</div>

## Composición de Funciones

<div style="text-align: justify; font-size:1.1rem;">

Para calcular $P$ y $L$ hemos aplicado una rotación $R^T$, luego multiplicamos por una matriz diagonal y luego por otra rotación $R$

$\Pi =
\begin{bmatrix}
0&0\\
0&1
\end{bmatrix}$

$ R = \begin{bmatrix}
\frac{3}{5} & -\frac{4}{5}\\
\frac{4}{5} & \frac{3}{5}\\
\end{bmatrix}$

$ F = \begin{bmatrix}
-1 & 0\\
0 & 1\\
\end{bmatrix}$

$P = R\Pi R^T$

$L = RFR^T$

La idea de esto es componer transformaciones lineales. Es decir, aplicar una a continuación de la otra. En términos de las matrices, la composición de transformaciones lineales implica el producto entre matrices. SOn a su vez ejemplos de factorizaciones de matrices.

Realizar cálculos por medio de la factorización de metrices, o composición de TLs, aporta mayor sencillez. 

Empecemos por la reflexión:

Primero necesitamos rotar nuestro sistemas de coordenadas para llevar los puntos del eje $x$ a la recta $4x - 3y = 0$. Eso es lo que hace $R^T$
</div>

````python
show_TL(R.T)
````

<p style="text-align: justify; font-size:1.1rem;">Luego realizamos una reflexión, con la matriz $F$. De hecho ya sabíamos como hacer eso cuando estudiamos $F$</p>

````python
show_TL(F @ R.T)
````

## Autovalores y autovectores

<div style="text-align: justify; font-size:1.1rem;">
A partir de este ejemplo, se nos desprende el concepto de <b>autovalores</b> (o eigenvalores o valores propios) y <b>autovectores</b> (o eigenvectores o vectores propios) de una matriz. <br>
Para tener una idea general, los autovectores de una matriz <i>cuadrada</i>, son aquellos vectores que multiplicados por la matriz en cuestión, nos darán como resultado ese mismo vector, que conserva la misma dirección del vector inicial, pero con una norma o longitud que puede ser mayor o menor. Ese aumento o disminución de la longitud nos la va a aportar el o los autovalores asociados a los distintos autovectores. Y estos autovalores nos son otra cosa que escalares (números reales o bien complejos).

En base a este concepto, tenemos entonces que se debe cumplir esta ecuación (y es la que nos permitiría encontrar los autovalores y autovectores de la matriz, como se demuestra en los ejemplos de los links de abajo):

La multiplicación de la matriz por un vector, nos da como resultado lo mismo que la multiplicación de un autovalor (escalar real o complejo) por ese mismo vector, que se denomina autovector <br>

$M . v = \lambda . v$   

Por otro lado, esta es la fórmula para encontrar el llamado <i>polinomio característico</i>, que luego nos permitirá hallar los autovalores y autovectores de la matriz: <br>

$det(M - \lambda.I) = 0$

## Diagonalización de una matriz

### Ejemplo de cálculo de autovalores y autovectores

<p style="text-align: justify; font-size:1.1rem;">Consideremos nuestra matriz $M$. Hasta ahora, alguien nos entrego sus autovalores y autovectores. Pero sería bueno saber como encontrarlos. </p>

#### Métodos iterativos para autovalores y autovectores

<div style="text-align: justify; font-size:1.1rem;">
    
Por ahora lo único que sabemos es que queremos encontrar un autovector $x_i$ y un autovalor $λ_i$ que verifican.

$Mx_i = \lambda_ix_i$

Supongamos que tenemos solo un vector inicial $v$ y apliquemos sucesivas veces $M$, es decir, el vector resultante d ela primer multiplicación lo vuelvo a multiplicar por $M$ (recordar que $M$ siempre está a la izquierda) y así sucesivamente.

Por ahora, esta es una forma un tanto casera o manual para ir encontrando autovectores:
</div>

````python
v = np.matrix([-2, 1]).T
print("itereación: 0", v[0],v[1])
for i in range(10):
  v = M @ v
  print("itereación:",i+1,v[0],v[1])
````

<div style="text-align: justify; font-size:1.1rem;">
    
Despues de iteraciones con un vector arbitrario obtenemos algo parecido a un autovector. Decimos que es algo parecido a un auto vector por lo siguiente: 

Sabíamos que $(x_1)^T = \begin{bmatrix} 1&0\end{bmatrix}$ era un autovector, pero tambien lo será $(y)^T = \begin{bmatrix} -2&0\end{bmatrix}$, $(u)^T = \begin{bmatrix} 100&0\end{bmatrix}$ o $(w)^T = \begin{bmatrix} -173&0\end{bmatrix}$. En nuestra iteración final, estamos cerca de un valor similar a $w$, un autovector propiamente dicho.

Vemos que las aplicaciones sucesivas sobre un vector nos acerca a un autovector. Con lo cual podemos proponer este truco para encontrar un autovector. De hecho podemos proponer una mejora. Dado que algunas componentes pueden aumentar (en este caso, la componente con valor 172...), es recomendable ir achicando nuestros vectores para evitar que estos crezcan indefinidamente. Lo mismo si nuestro vectores se achicaran. Para solucionar esto, eligiremos dividir la mayor componente en valor absoluto. Por ejemplo en nuestro vector $(v)^T = \begin{bmatrix}-2&1\end{bmatrix}$ deberiamos dividir todo por $-2$
</div>

````python
v = np.matrix([-2, 1]).T
print("itereación: 0", v[0],v[1])
for i in range(10):
  v = M @ v
  v_abs = abs(v) # aplicamos valor absoluto componente a componente
  n = v_abs.argmax() # elegimos la componente con mayor valor absoluto
  lamb = v[n].copy() # guardamos la componente con mayor valor absoluto  
  v /= lamb #dividimos por la componente con mayor valor absoluto, en este caso se aplica broadcasting  
  print("itereación:",i+1,v[0],v[1],lamb,n)
````

<div style="text-align: justify; font-size:1.1rem;">
    
Al guardar la componente con mayor valor absoluto y dividir por ella, en esencia hemos encontrado también un autovalor.

La elección de dividir por la componente en mayor valor absoluto no es arbitraria. Puede demostrarse que nuestro algoritmo converge a un autovector si dividimos nuestros resultados por cualquier norma de $v$. En ese sentido, podríamos haber dividido por cualquier norma. Podríamos haber dividido por la norma euclidiana para encontrar los autovectores. Sin embargo, si elegimos la llamada **norma infinito**, esta norma nos devuelve el valor absoluto de el autovalor correpondiente. De hecho al quedarnos con la componente original, recuperamos el signo.

NormaInfinito($v$):
  1. calculo el valor absoluto de cada componente de $v$
  2. busco $k$, el máximo de los valores absolutos 
  3. devuelvo el valor $k$ encontrado.

```
def NormaInfinito(v):
  u = v.abs()
  k = u.max()
  return k
```
</div>

<div style="text-align: justify; font-size:1.1rem;">

Podemos ver que algoritmo nos ha permitdo calcular un solo autovector y un solo autovalor. De hecho, puede demostrarse que siempre encuentra el autovalor con mayor valor absoluto y autovector asociado.

Sin embargo, aun nos faltan encontrar otros autovalores y autovectores. Vemos que este problema puede ser muy intensivo en cálculo, si seguimos por este camino.

Afortunadamente, existen técnicas númericas más avanzadas inspiradas en este algoritmo. Estás técnicas pemiten encontrar todos los autovalores y autovectores. Además, estás ya se encuentran incluidas en `numpy`
</div>

````python
eigenvalues, eigenvectors = np.linalg.eig(M) #cálculo de autovectores y autovalores proporcionado por Numpy
print(eigenvalues[0], eigenvectors.T[0]) # Primer autovalor con su/s autovector/es
print(eigenvalues[1], eigenvectors.T[1]) # Segundo autovalor con su/s autovector/es
````

<p style="text-align: justify; font-size:1.1rem;">En efecto, a la salida tenemos cada autovalor y su autovector correspondiente. Debido a como se trabaja el algoritmo, los autovectores siempre terminan normalizados con norma euclidiana.</p>

### Diagonalización

<div style="text-align: justify; font-size:1.1rem;">

Antes de abordar el tema de la diagonalización de una matriz en los ejemplos de abajo, que está muy vinculada a los autovalores y autovectores, aquí quedan unos links con la explicación conceptual y ejemplos de este tema. Tener en cuenta, además, que diagonalizar una amtriz es un tipo de factorización (es decir, expresar una matriz como el producto de otras matrices), pero que **SOLO** aplica para matrices **cuadradas**
<br>
<br>
<a href="https://www.youtube.com/watch?v=-duROaTigUE" target="_blank">MateFacil - Diagonalización de una Matriz en 4 pasos</a>
<br>
<a href="https://www.youtube.com/watch?v=M6sxDXVRJXE" target="_blank">MateFacil - Diagonalización de una Matriz 3x3</a>

Ahora sí, pasamos a los apuntes propios de esta clase:
</div>

<div style="text-align: justify; font-size:1.1rem;">

Definamos las siguientes matrices a partir de los autovectores y autovalores de $M$ arbitraria:

$D = \begin{bmatrix}
λ_1&0&\dots & 0\\
0&λ_2&\dots & 0\\
\vdots&\vdots&\ddots & \vdots\\
0&0&\dots & λ_n\\
\end{bmatrix}$

$P = 
\begin{bmatrix}
p_1|p_2|\dots |p_n
\end{bmatrix}$

Es decir, en $D$ ponemos todos los autovalores $d_i$ de $M$ en la diagonal. El resto de los elementos los llenamos de 0. Por otro lado en $P$ ponemos a los autovectores $p_i$ de $M$ como columnas, **respetando que el orden de los auto vectores coindida con los autovalores**. 

Se puede demostrar que **si $P$ es invertible**, entonces:

$M = PDP^{-1}$

$PDP^{-1}$ es la composición de aplicar  $P^{-1}$, luego $D$ y finalmente $P$. Para nuestra $M$ tenemos:
</div>

<div style="text-align: justify; font-size:1.1rem;">

$M = PDP^{-1}$

$M = \begin{bmatrix}
-\frac{3}{2} & 2\\
0 & \frac{1}{2}\\
\end{bmatrix},~
D = \begin{bmatrix}
-\frac{3}{2} & 0\\
0 & \frac{1}{2}\\
\end{bmatrix}$

$ P^{-1} =
\begin{bmatrix}
1 & -1\\
0 & \sqrt{2}\\
\end{bmatrix},~ 
P = \begin{bmatrix}
1 & \dfrac{\sqrt{2}}{2}\\
0 & \dfrac{\sqrt{2}}{2}\\
\end{bmatrix}$

Claramente, la diagonalización es un ejemplo de factorización
</div>

````python
P = eigenvectors # Los autovectores que antes calculamos de la matriz M
P_inv = np.linalg.inv(P)
print("Matriz P de autovectores")
print(P)
print()
print("Inversa de P")
print(P_inv)
print()

D = np.zeros(M.shape)
for i in range(eigenvalues.size):
  D[i][i] = eigenvalues[i]
print("Matriz D diagonal de autovalores")
print(D)
print()

print("Matriz P.D.P^-1 es igual a matriz M")
print(P @ D @ P_inv)
print()
print(M)
````

<p style="text-align: justify; font-size:1.1rem;">En función a lo anterior podemos ver que hay matrices, como $M$, que pueden ser representadas como diagonales. Para que puedan ser representadas como diagonales, necesitamos que esta matriz $P$ de autovectores sea invertible.</p>

#### La diagonalización no es única

<div style="text-align: justify; font-size:1.1rem;">

$M = P^{-1}DP = Y^{-1}DY$

$M = \begin{bmatrix}
-\frac{3}{2} & 2\\
0 & \frac{1}{2}\\
\end{bmatrix},~
D = \begin{bmatrix}
-\frac{3}{2} & 0\\
0 & \frac{1}{2}\\
\end{bmatrix}$

$ Y^{-1} =
\begin{bmatrix}
1 & 1\\
0 & 1\\
\end{bmatrix},~ 
Y = \begin{bmatrix}
1 & -1\\
0 & 1\\
\end{bmatrix}$

$ P^{-1} =
\begin{bmatrix}
1 & -1\\
0 & \sqrt{2}\\
\end{bmatrix},~ 
P = \begin{bmatrix}
1 & \dfrac{\sqrt{2}}{2}\\
0 & \dfrac{\sqrt{2}}{2}\\
\end{bmatrix}$
</div>

````python
P = eigenvectors
P_inv = np.linalg.inv(P)

Y = np.matrix([[1,1],
               [0,1]])
Y_inv = np.linalg.inv(Y)

D = np.zeros(M.shape)
for i in range(eigenvalues.size):
  D[i][i] = eigenvalues[i]

print(P @ D @ P_inv)
print(Y @ D @ Y_inv)
print(M)
````

#### No toda matriz es diagonalizable 

<div style="text-align: justify; font-size:1.1rem;">

* Las matrices que representan rotaciones son diagonalizables... *si usamos números complejos como autovalores*

* Una deformación de corte no puede diagonalizarse. Para esos casos se usan otras herramientas que generalizan noción de *diagonalización* (Formas de Jordan)
</div>

#### Diagonalización de matrices simétricas

<div style="text-align: justify; font-size:1.1rem;">

Toda matriz simétrica $B$ es diagonalizable
* Todos los autovalores $d_i$ son *reales*
* Los autovectores $p_i$ forman una base ortogonal.
 
  Es decir, si usamos autovectores con norma unitaria para construir $P$ se verifica:

  $P^T=P^{-1}$

* Equivalentemente $P$ representa una rotación
</div>

````python
B = np.matrix([[1,2],
               [2,3]])
# eigh es una función optimizada para matrices simétricas (lo mismo que eig() pero para matrices simetricas)
B_values, B_vectors = np.linalg.eigh(B)

B_diag = np.zeros(B.shape)
for i in range(B_values.size):
  B_diag[i][i] = B_values[i]

print(B)
print()
print(B_diag)
print(B_vectors)
print()
print(B_vectors @ B_vectors.T)
print(B_vectors @ B_diag @ B_vectors.T)
````

## Descomposición en valores singulares

<div style="text-align: justify; font-size:1.1rem;">

La descomposición en valores singulares de una matriz (o SVD por sus siglas en inglés, <i>Single Values Decomposition</i>), es también un tipo de factorización de una matriz, que en cierta forma se parece a la diagonalización, pero a diferencia de esta última, sirve para todo tipo de matrices (cuadradas y no cuadradas).

Su fórmula se expresa de esta manera:

$M = U\sigma V^{-1}$  

* $\sigma$: es una matriz del mismo tamaño que $M$ y en la diagonal principal (si es rectangular, la diagonal principal empieza en la fila 1 - columna 1 siempre), se encuentran las raíces cuadradas de los autovalores de la matriz simétrica y cuadrada que se obtiene de $M$.$M^T$, ordenados de mayor a menor. En el resto de las posiciones, solo hay ceros. Estos valores de $\sigma$ en la diagonal son lo que se conoce como **valores singulares** de $M$
* $U$ y $V^{-1}$: son matrices cuadradas, ortogonales (su inversa es igual a su transpuesta) que se obtienen a partir de los autovectores de $M$.$M^T$
* $U$: si $M$ es una matriz de $m$ x $n$, $U$ es una matriz de $m$ x $m$ (toma la cantidad de filas de $M$). Cada vector columna de esta matriz se obtiene mediante el cálculo de la multiplicación entre: valor singular x matriz $M$ x autovector normalizado (emergido del autovalor del que proviene el valor singular). Esto está más claro en el video de abajo
* $V$: si $M$ es una matriz de $m$ x $n$, $V$ es una matriz de $n$ x $n$ (toma la cantidad de columnas de $M$). Cada vector columna de esta matriz se forma con los autovectores obtenidos de $M$.$M^T$, pero normalizados. Y si no alcanzaran los autovectores encontrados, se deben buscar por medio del método de base del subespacio ortogonal (ver link de abajo), y luego normalizando el vector extra hallado.
