### Step 14. Present only the Shooting Accuracy from England, Italy and Russia

```python
def contains_country(countries, country):
  for c in countries:
    if country.lower() == c.lower():
      return True
  return False
```

```python
countries = {"England", "Italy", "Russia"}

euro12[euro12["Team"].apply(lambda x: contains_country(countries, x))] \
      ["Shooting Accuracy"]
```
